```python
#Transformar los archivos netflix_movies.xlsx y turismo_post_pandemia.xlsx a csv 
```

```python
#Importar libreria adicional
import json
```

```python
# 1. netflix_movies.xlsx
```

```python
#DataFrame creado con un archivo xlsx
df2 = pd.read_excel("netflix_movies.xlsx")
df2
```

```python
#DataFrame transformado a un diccionario de formato json
data_json = df2.to_dict(orient='records')
data_json
```

```python
#Guardar el diccionario como json en un archivo
with open("netflix_movies.json",'w') as json_file:
    json.dump(data_json,json_file)
```
